```python
from scipy.stats import norm
import numpy as np

annual_mean = 0.08  # 8%
annual_sd = 0.04  # 4%
investment = 500e6  # $500 million

# Quarterly
quarterly_mean = annual_mean / 4
quarterly_sd = annual_sd / np.sqrt(4)
# Monthly
monthly_mean = annual_mean / 12
monthly_sd = annual_sd / np.sqrt(12)
# Daily
daily_mean = annual_mean / trading_day
daily_sd = annual_sd / np.sqrt(trading_day)

# Probabilities of a negative return
# Annualy
prob_negative_annual = norm.cdf(0, annual_mean, annual_sd)
# Quarterly
prob_negative_quarterly = norm.cdf(0, quarterly_mean, quarterly_sd)
# Monthly
prob_negative_monthly = norm.cdf(0, monthly_mean, monthly_sd)
# Daily
prob_negative_daily = norm.cdf(0, daily_mean, daily_sd)

expected_pnl_quarterly = investment * quarterly_mean
pnl_sd_quarterly = investment * quarterly_sd
pnl_range_quarterly = pnl_sd_quarterly * 2 


print('The probability that BCM will lose money in given year is '+ "{:.3f}".format(prob_negative_annual))
print('The probability that BCM will lose money in given quarter is '+ "{:.3f}".format(prob_negative_quarterly))
print('The probability that BCM will lose money in given month is '+ "{:.3f}".format(prob_negative_monthly))
print('The probability that BCM will lose money in given day is '+ "{:.3f}".format(prob_negative_daily))
print('The expected earning at the first quarter is ' + "{:.3f}".format(expected_pnl_quarterly))
print('The the +/- 2 standard deviation is ('+"{:.3f}".format(expected_pnl_quarterly-pnl_range_quarterly)+','+"{:.3f}".format(expected_pnl_quarterly+pnl_range_quarterly)+')')


```

```python

std_deviations = np.arange(2, 12.25, 0.25) / 100 

annual_return = 0.08
daily_return = annual_return / trading_day  
constant_annual_sd = 0.04  
daily_sd = constant_annual_sd / np.sqrt(trading_day)

variation_sd_results = {
    'Standard Deviation': [],
    'Winning Rate': [],
    'Sharpe Ratio': [],
    'Up-Down Ratio':[]
}

def calc_win_rate_sharpe(sd_array, expected_return):
    win_rate = []
    sharpe_ratio = []
    for sd in sd_array:
        daily_sd_varying = sd / np.sqrt(trading_day)
        win_rate.append(1 - norm.cdf(0, daily_return, daily_sd_varying))
        sharpe_ratio.append(expected_return / sd)
    
    return win_rate, sharpe_ratio

def calc_up_down_ratio(daily_mean, daily_sd):
    expected_up = (1 - norm.cdf(daily_mean, daily_mean, daily_sd)) * daily_mean
    expected_down = norm.cdf(-daily_mean, daily_mean, daily_sd) * daily_mean
    up_down_ratio = expected_up / abs(expected_down)
    return up_down_ratio

winning_rates_sd, sharpe_ratios_sd = calc_win_rate_sharpe(std_deviations, annual_return)
variation_sd_results['Standard Deviation'] = std_deviations
variation_sd_results['Winning Rate'] = winning_rates_sd
variation_sd_results['Sharpe Ratio'] = sharpe_ratios_sd

up_down_ratios_sd = []
for sd in std_deviations:
    daily_sd_varying = sd / np.sqrt(trading_day)
    up_down_ratio = calc_up_down_ratio(daily_return, daily_sd_varying)
    up_down_ratios_sd.append(up_down_ratio)

variation_sd_results['Up-Down Ratio'] = up_down_ratios_sd

variation_sd_results_df = pd.DataFrame(variation_sd_results)
variation_sd_results_df
```

```python
expected_returns = np.arange(2, 12.25, 0.25) / 100  
variation_return_results = {
    'Expected Return': [],
    'Winning Rate': [],
    'Sharpe Ratio': [],
    'Up-Down Ratio': []
}

daily_sd_constant = daily_sd

for expected_return in expected_returns:
    daily_return_varying = expected_return / np.sqrt(trading_day)
    winning_rate = 1 - norm.cdf(0, daily_return_varying, daily_sd_constant)
    sharpe_ratio = expected_return / constant_annual_sd
    up_down_ratio = calc_up_down_ratio(daily_return_varying, daily_sd_constant)
    variation_return_results['Expected Return'].append(expected_return)
    variation_return_results['Winning Rate'].append(winning_rate)
    variation_return_results['Sharpe Ratio'].append(sharpe_ratio)
    variation_return_results['Up-Down Ratio'].append(up_down_ratio)

variation_return_results_df = pd.DataFrame(variation_return_results)
variation_return_results_df
```

## iii)

- Analysis
    - The Winning Rate and Up-Down Ratio can reflect the short-term performance of a strategy, while the Sharpe Ratio provides a risk-adjusted evaluation of the strategy's longer-term performance.

    - Sharpe Ratio:
        - measures the excess return per unit of risk, which is the risk-adjusted return. When holding standard deviation (risk) constant, an increase in expected return leads to a higher Sharpe Ratio; conversely, a decrease in expected return results in a lower Sharpe Ratio.
        - Risk-Reward Tradeoff: This fundamental economic principle states that higher risks should be compensated with higher potential rewards. The Sharpe Ratio encapsulates this principle by comparing the additional return of an investment over the risk-free rate to the additional risk taken (standard deviation). A higher Sharpe Ratio indicates a more favorable risk-reward tradeoff.
        - Efficient Market Hypothesis (EMH): In an efficient market, all known information is already reflected in prices, and higher returns can only be achieved by taking on more risk. The Sharpe Ratio is a measure that reflects how well an investment performs in excess of the risk-free rate, relative to its added risk.
    - Winning Rate:
        - the proportion of days with positive returns within a given period. As the standard deviation increases, even with a constant expected return, the winning rate decreases due to higher uncertainty in returns; conversely, an increase in the expected return rate, with risk constant, results in a higher winning rate because the likelihood of positive returns increases.
        - Prospect Theory: Investors value gains and losses differently, leading to decisions based on perceived gains rather than actual outcomes. A higher Winning Rate might be preferred by investors due to the preference for frequent positive feedback, despite the magnitude of gains or losses.
        - Behavioral Biases: The Winning Rate could be a metric that appeals to the investor’s behavioral biases like overconfidence where investors might prefer strategies that win more often, underestimating the impact of the magnitude of gains or losses.
    - Up-Down Ratio:
        - Utility Theory: This assumes investors prefer higher returns with less risk. The Up-Down Ratio can be seen as a utility function where higher average returns on winning days compared to losses on losing days (regardless of the frequency of such days) would be preferred by a rational investor.
        - Liquidity Preference Theory: This could explain preferences for strategies with higher Up-Down Ratios as they might be associated with more liquid investments, allowing investors to enter and exit positions more favorably.

- Assumptions
    - Rational Behavior: Investors are assumed to make rational decisions that maximize their utility.
    - Risk Aversion: Most investors are assumed to be risk-averse, meaning they prefer a less uncertain return. The Sharpe Ratio is particularly relevant under this assumption, as it informs investors about how much excess return is generated per unit of risk.
    - Information Symmetry: The calculations assume that all investors have access to the same information and will respond similarly to that information. In reality, information asymmetry can exist.
    - Market Liquidity: For the Up-Down Ratio, there is an assumption of sufficient market liquidity that allows investors to realize their gains or cut their losses on any given day.

```python
from scipy.stats import norm
import numpy as np

def simulate_drawdowns(daily_mean, daily_sd, period_days, drawdown_length):
    daily_returns = np.random.normal(daily_mean, daily_sd, period_days)
    fund_value = 10000 * np.cumprod(1 + daily_returns)
    
    peak = fund_value[0]
    drawdown_counter = 0
    drawdown_occurred = False
    
    for value in fund_value:
        if value > peak:
            peak = value
            drawdown_counter = 0
        else:
            drawdown_counter += 1
            if drawdown_counter >= drawdown_length:
                drawdown_occurred = True
                break
    
    return drawdown_occurred


annual_mean = 0.08
annual_sd = 0.04
period_years = 5
period_days = period_years * trading_day
drawdown_length = 90 

daily_mean = annual_mean / trading_day
daily_sd = annual_sd / np.sqrt(trading_day)

num_simulations = 10000  

drawdown_occurrences_90 = [simulate_drawdowns(daily_mean, daily_sd, period_days, 90)
                        for _ in range(num_simulations)]

probability_drawdown_90 = np.mean(drawdown_occurrences_90)

drawdown_occurrences_120 = [simulate_drawdowns(daily_mean, daily_sd, period_days, 120)
                        for _ in range(num_simulations)]

probability_drawdown_120 = np.mean(drawdown_occurrences_120)
print('The probability for holding 5 years to observe 90 day drawdown is '+"{:.3f}".format(probability_drawdown_90))
print('The probability for holding 5 years to observe 120 day drawdown is '+"{:.3f}".format(probability_drawdown_120))
```

And if we Lengthening the Investment Period, the probability of 90-day drawdown observation would increase, potentially approaching certainty (100%), because given more time, the likelihood of experiencing every possible market scenario, including drawdowns, increases. And for 120-day drawdown, the probability would also increase for the same reasons, but it might not approach certainty as quickly as the 90-day drawdown probability due to the longer recovery time required.

```python
import pandas as pd


trading_days = 252
standard_deviation = 0.04  
risk_free_rate = 0.0  
initial_investment = 500e6  
sharpe_ratios = np.arange(-1, 3.5, 0.5)  
drawdown_lengths = [1, 3, 5, 10, 20, 40, 60, 90, 120]  
num_simulations = 10000  


def simulate_drawdowns_for_sharpe_ratio(sharpe_ratio, sd, risk_free_rate, drawdown_lengths, num_simulations, trading_days):

    expected_return = risk_free_rate + sharpe_ratio * sd
    daily_mean = expected_return / trading_days
    daily_sd = sd / np.sqrt(trading_days)
    drawdown_observations = {length: 0 for length in drawdown_lengths}
    
    for _ in range(num_simulations):
       
        daily_returns = np.random.normal(daily_mean, daily_sd, trading_days)
        fund_value = initial_investment * np.cumprod(1 + daily_returns)
        peak_value = fund_value[0]
        drawdown_counter = 0  
        flag = {day: 0 for day in drawdown_lengths}

        for value in fund_value:
            if value > peak_value:
                # New peak, so reset drawdown counter
                drawdown_counter = 0
                peak_value = value
            else:
                drawdown_counter += 1
                for length in drawdown_lengths:
                    if (drawdown_counter == length)&(flag[length]!=1):
                        drawdown_observations[length] += 1
                        flag[length] = 1
                        break 
    
    drawdown_probabilities = {length: count / num_simulations for length, count in drawdown_observations.items()}
    return drawdown_probabilities

drawdown_table = {sharpe_ratio: simulate_drawdowns_for_sharpe_ratio(sharpe_ratio, standard_deviation, risk_free_rate, drawdown_lengths, num_simulations, trading_days) for sharpe_ratio in sharpe_ratios}


drawdown_df = pd.DataFrame(drawdown_table)
drawdown_df
```

This period should include different market environments to allow the manager’s decisions to be tested under various conditions. Short-term results can be influenced by market noise or temporary trends, whereas a full cycle would give a clearer picture of the manager's skill.

- Learning About Skill from Winning Rates and Up-Down Ratios:
    - Yes, these ratios can be indicative of skill. A consistently high winning rate suggests that the manager can generate positive returns more frequently. A favorable up-down ratio indicates the manager is able to capitalize more on winning days than they lose on down days, suggesting good decision-making skills and effective risk management. However, these should be considered alongside other metrics to understand the full picture.

- Limits of Arbitrage:
    - The limits of arbitrage theory suggest that even if mispricings are identified, it may not always be possible to profit from them due to costs, risks, and constraints. Managers with skill in identifying and exploiting these opportunities will have a higher SR, but the table indicates that longer-term drawdowns can still occur. This underlines the necessity for robust risk management in the face of arbitrage limitations.

- Fundamental Law of Active Management:
    - This law posits that a manager's performance is a function of skill and breadth of independent bets. The table reinforces this law; to maintain a high SR, a manager needs not only the skill but also a sufficient number of opportunities to apply this skill. This reduces the influence of luck and emphasizes the importance of skill in achieving sustained success.

- Frequency of Reporting(Greater Transparency):
    - Pros: Frequent reporting keeps investors informed and may build trust. It allows for ongoing scrutiny of fund performance and manager skill, potentially leading to increased investment if the fund performs well.
